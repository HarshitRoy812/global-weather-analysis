### DATA CLEANING

```python
weather_df['last_updated'] = pd.to_datetime(weather_df['last_updated'])
```

```python
weather_df['moonrise'].str.contains('No moonrise').any()
```

```python
# There is some time where the moon doesn't rise as well as doesn't set so drop those rows

weather_df.drop(weather_df[weather_df['moonrise'] == 'No moonrise'].index,inplace = True)
weather_df.drop(weather_df[weather_df['moonset'] == 'No moonset'].index,inplace = True)
```

```python
weather_df['moonrise'].str.contains('No moonrise').any()
```

```python
weather_df['moonset'].str.contains('No moonset').any()
```

```python
import datetime

def convertToTime(obj):
    time_obj = datetime.datetime.strptime(obj,'%I:%M %p')
    converted_obj = time_obj.strftime('%H:%M')

    return converted_obj

weather_df['sunset'] = weather_df['sunset'].apply(convertToTime)
weather_df['sunrise'] = weather_df['sunrise'].apply(convertToTime)
weather_df['moonrise'] = weather_df['moonrise'].apply(convertToTime)
weather_df['moonset'] = weather_df['moonset'].apply(convertToTime)
```

```python
# Dropping those columns that are not required for our analysis

weather_df.drop(columns = ['wind_mph','moon_illumination','moon_phase','air_quality_gb-defra-index','air_quality_us-epa-index','air_quality_PM10','air_quality_PM2.5',
                   'gust_mph','visibility_miles','last_updated_epoch'],inplace = True)
```

```python
# Renaming columns for better readability and understanding

weather_df.rename(columns = {'location_name' : 'location','last_updated' : 'date','temperature_celsius' : 'temp_celsius','temperature_farenheit' : 'temp_farenheit',
                             'condition_text' : 'condition','air_quality_Carbon_Monoxide' : 'air_quality_co','air_quality_Ozone' : 'air_quality_ozone',
                             'air_quality_Nitrogen_dioxide' : 'air_quality_no2','air_quality_Sulphur_dioxide' : 'air_quality_so2'},inplace = True)
```

```python
weather_df['condition'].unique()
```

```python
# Now as the 'condition' column of the dataframe has many values which are not required for analysis, hence we will replace them with a singular value for
## any matching conditions e.g -> (showers,drizzle) = Rain

def replaceWithMatch(str):

    temp = str.lower()

    if 'rain' in temp or 'showers' in temp or 'shower' in temp or 'drizzle' in temp or 'thundery' in temp or 'thunder' in temp:
        return 'Rain'

    if 'snow' in temp or 'sleet' in temp or 'ice' in temp or 'blizzard' in temp:
        return 'Snow'

    if 'fog' in temp:
        return 'Fog'

    return str.title().strip()
        
weather_df['condition'] = weather_df['condition'].apply(replaceWithMatch)
```

### EXPLORATORY DATA ANALYSIS

#### Lets visualize which countries experience the most rainfall

```python
temp_df = weather_df[weather_df['condition'] == 'Rain']

rainfall = temp_df['country'].value_counts()
country = temp_df['country'].unique()

plt.figure(figsize = (10,32))
bar_plot = sns.barplot(x = rainfall,y = country,color = '#0d999e');
plt.xlabel('Rainfall')
plt.ylabel('Country')
plt.xticks([])

plt.show()
```

##### From the above chart, we can say that ****Antigua and Barbuda**** and ****Austria**** are the top 2 countries that experience high rainfall and countries like ****Egypt**** and ****Angola**** experience the lowest rainfall due to its deserty conditions

#### Plotting correlation matrix heatmap to see any correlation between variables

```python
num_cols = ['temp_celsius','wind_kph','wind_degree','pressure_in','precip_in','visibility_km','uv_index','air_quality_co']

plt.figure(figsize = (12,10))
sns.heatmap(weather_df[num_cols].corr(),annot = True,cmap = 'viridis')
plt.show()
```

##### We can visualize from the above heatmap that there is some positive correlation between ****Temperature**** and ****UV Index**** but we cannot say that it high temperature may cause high uv index or vice-versa

##### There is also some slight positive correlation between ****UV Index**** and ****Wind Speed**** but cannot be the causation

#### Visualizing the rainfall in India over a certain period of time

```python
india_df = weather_df[(weather_df['country'] == 'India') & (weather_df['condition'] == 'Rain')]
india_df.set_index('date',inplace = True)

# Resampling the df by 'Month' to get total precipitation (mm) for a particular month
india_df = india_df.resample('M').sum()

date = india_df.index
rainfall = india_df['precip_mm']

sns.lineplot(x = date,y = rainfall,color = '#27a85b',marker = 'o')
plt.title('Monthly Rainfall(mm) in India')
plt.show()
```
